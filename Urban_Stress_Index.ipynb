##Feature Engineering


```python
df['pollution_stress'] = (
    df['avg_pm25'] * 0.35 +
    df['avg_pm10'] * 0.25 +
    df['avg_no2'] * 0.15 +
    df['avg_o3'] * 0.15 +
    df['avg_so2'] * 0.10
)
```

```python
df['pollution_stress'].head(5)
```

```python
df['heat_stress'] = df['temperature'] * (df['humidity'] / 100)
```

```python
df['heat_stress'].head(5)
```

```python
df['infra_density'] = df['total_infrastructure'] / df['population']
```

```python
df['infra_density'].head(5)
```

```python
df['green_space_per_capita'] = df['total_green_area_sqm'] / df['population']
```

```python
df['green_space_per_capita'].head(5)
```

```python
from sklearn.preprocessing import MinMaxScaler

features_to_scale = [
    'pollution_stress',
    'heat_stress',
    'infra_density',
    'green_space_per_capita'
]

scaler = MinMaxScaler()
df[features_to_scale] = scaler.fit_transform(df[features_to_scale])
```

```python
df[['pollution_stress','heat_stress','infra_density','green_space_per_capita']].describe()
```

```python
df['urban_stress_index'] = (
    df['pollution_stress'] * 0.35 +
    df['heat_stress'] * 0.25 +
    df['infra_density'] * 0.25 -
    df['green_space_per_capita'] * 0.15
)
```

```python
df['urban_stress_index'].head(5)
```

##ML Modeling

```python
ml_features = [
    'pollution_stress',
    'heat_stress',
    'infra_density',
    'green_space_per_capita'
]
```

```python
ml_features
```

```python
X = df[ml_features]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []

for k in range(2, 8):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(range(2, 8), inertia, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42)
df['stress_cluster'] = kmeans.fit_predict(X_scaled)
```

```python
cluster_summary = df.groupby('stress_cluster')[ml_features].mean()
cluster_summary
```

```python
cluster_labels = {
    0: "Very Low Stress",
    1: "Low Stress",
    2: "Moderate Stress",
    3: "High Stress",
    4: "Extreme Stress"
}

df['stress_level_ml'] = df['stress_cluster'].map(cluster_labels)
```

```python
cluster_summary.sort_values(by='stress_cluster')
```

```python
df[['stress_cluster', 'stress_level_ml']].head(10)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='stress_level_ml', order=["Very Low Stress","Low Stress","Moderate Stress","High Stress","Extreme Stress"])
plt.title("Number of Cities per Stress Level")
plt.xticks(rotation=0)
plt.show()
```

```python
features = ['urban_stress_index', 'avg_pm25', 'population', 'num_green_spaces', 'temperature']

df.groupby('stress_level_ml')[features].mean().round(2)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
components = pca.fit_transform(df[['population','avg_pm25','temperature','total_infrastructure','num_green_spaces']])

df['pca1'] = components[:,0]
df['pca2'] = components[:,1]

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='pca1', y='pca2', hue='stress_level_ml', palette='coolwarm')
plt.title("City Stress Levels in PCA 2D")
plt.show()
```
